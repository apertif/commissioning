This notebook looks at the variations in amplitude and phase vs. time and frequency for a calibrator that is placed at the HPFW of a beam. It does this using the Miriad task "uvspec" to shift the phase center and extract a spectrum.

This example notebook lives in `commissioning/beam_stability` but additional tests using it should be run in `commissioning/beam_stability/testing`

This example notebook is running on WSRTA180608006_B019.MS, which has 3C48 at the HPFW.

This is uncalibrated data - there were not separate calibration observations for this source. In the future, it is probably better to do a calibration observation (calibrator in center of beam) so that anything that is a calibration issue is not included.

```python
#set up apercal environment and logger

import os;
myusername = os.environ['USER']
if not ('PYTHONPATH' in os.environ and
        '/home/' + myusername + '/apercal' in os.environ['PYTHONPATH']):
      print("WARNING: your apercal directory should be in the $PYTHONPATH")

%config IPCompleter.greedy=True
%matplotlib notebook
import libs.lib as lib
lib.setup_logger('info', logfile='/home/{}/commissioning/beam_stability/testing/logs/cal_hpfw_uncal.log'.format(myusername))

import apercal
```

This notebook starts with raw data. We will not calibrate it, but we do still need to flag and convert it, so we will use that part of apercal. Note that in this case, the data was retrieved manually from ALTA and placed in the correct spot, so we skip the prepare step. 

```python
#globally define the cfgfile
cfgfile = '/home/{}/commissioning/beam_stability/testing/cfg/cal_hpfw_uncal.cfg'.format(myusername)
```

```python
#Do manual flags on the data
#With no calibrator obs, can't do bandpass correction
#Therefore can't run aoflagger (expects corrected column)

preflag = apercal.preflag(cfgfile)
nchannel = 11008
a = range(0, nchannel, 64) # the subband edges
b = range(1, nchannel, 64)
c = range(63, nchannel, 64)
d = range(16, nchannel, 64) # the two ghosts
e = range(48, nchannel, 64)

l = a + b + c + d + e
preflag.preflag_manualflag_channel = ';'.join(str(ch) for ch in l)

preflag.manualflag()
```

```python
#And convert
#without calibrator for bandpass, can't do aoflagger
convert = apercal.convert(cfgfile)
convert.go()
```

Now that we have a miriad (*uncalibrated*) dataset, we want to see how the flux and phase vary over time for a source as compared to its average value.  First we need the coordinates of a source, and the beam phase center.  

```python
filename=preflag.target.replace('MS','mir')
uvfile=(preflag.basedir+str(preflag.beam)+'/crosscal/'+filename)
print(uvfile)

# Use Miriad task to extract RA, Dec of phase center.
uvlist=lib.miriad('uvlist')
uvlist.vis=uvfile
uvlist.options='source'
uvlist.log='/home/{}/apercal/temp/mirlog/source.log'.format(myusername)
uvlist.go()

# Extract the RA, Dec of the beam pointing center
with open('/home/{}/apercal/temp/mirlog/source.log'.format(myusername)) as f:
    tmp_text=f.readlines()
f.close()
RAc=tmp_text[6][5:16]
Decc=tmp_text[7][5:16]

RAc=RAc.replace(':','h',1)
RAc=RAc.replace(':','m',2)
RAc=RAc+'s'

Decc=Decc.replace(':','d',1)
Decc=Decc.replace(':','m',2)
Decc=Decc+'s'

# Convert to SkyCoord
from astropy.coordinates import SkyCoord  # High-level coordinates

ptg=SkyCoord(RAc,Decc)

print(ptg)

#clean up and remove tmp log file
```

```python
from astropy import units as u

# Enter the RA,Dec of the source of interest as a string in the format below:
#RAs='00h42m18.368s'
#Decs='41d29m22.85s'
RAs='00h42m18.368s'
Decs='41d29m22.85s'

src=SkyCoord(RAs,Decs)

# Calculate the difference between the positions in arcsec
dRA,dDec=ptg.spherical_offsets_to(src)

dRA=dRA.to(u.arcsec).value
dDec=dDec.to(u.arcsec).value

offsets="'{0},{1}'".format(dRA,dDec)

print(offsets)
```

Now we will pick some averaging time intervals

```python
#Enter the solution time in minutes here.
tsolve=1000.   # This is the timescale over which the reference spectrum is calculated. 

#Enter the time scale over which you'd like to examine variations in amplitude and phase
tint=30.  # This is in minutes
```

Now we can use UVSPEC to extract an average spectrum for the entire observation.  

```python
uvspec = lib.miriad('uvspec')

uvspec.vis = uvfile
uvspec.interval = tsolve  # Plot over the entire solution interval
uvspec.stokes='ii'
uvspec.axis='frequency,amplitude'
uvspec.options='nobase,avall'
uvspec.log='/home/{}/testing/logs/contsrc_12h.log'.format(myusername)
uvspec.offset=offsets

uvspec.go()
```

Now we will load the average bandpass into an array, plot it, and save it for reference. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

freq_avg=np.loadtxt('/home/{}/testing/logs/contsrc_12h.log'.format(myusername),float,usecols=[0])
amp_avg=np.loadtxt('/home/{}/testing/logs/contsrc_12h.log'.format(myusername),float,usecols=[1])
nchan=len(freq_avg)
plt.plot(freq_avg,amp_avg,'-')
plt.xlabel('Frequency [GHz]')
plt.ylabel('Amplitude [Jy]')
plt.show()
```

Okay, now let's run uvspec with a shorter averaging interval and save the results to another log file, this time with a timestamp in it.  

```python
uvspec = lib.miriad('uvspec')

uvspec.vis = uvfile
uvspec.interval = tint  #Averaging the calibrator over tint.
uvspec.stokes='ii'
uvspec.axis='frequency,amplitude'
uvspec.options='nobase,avall,timelog'
uvspec.log='/home/{}/testing/logs/contsrc_tint.log'.format(myusername)
uvspec.offset=offsets

uvspec.go()
```

Okay, now we need to parse this new log file so that we can extract each time interval into a separate array.  To do this, we will use the original extracted spectrum as a guide to the number of channels we need to extract.  

We also need to know the number of intervals, which we can do with the following code:

```python
def lcount(keyword, fname):
    with open(fname, 'r') as fin:
        return sum([1 for line in fin if keyword in line])
```

```python
nint=lcount(':','/home/{}/testing/logs/contsrc_tint.log'.format(myusername))
print("Number of time intervals: ",nint)
print("Number of channels in spectrum: ", nchan)
```

We can use grep on the uvspec_avg.log file to extract the exact times corresponding to each interval as well.

We will now use this information to parse the log file, saving each time interval to a separate log.

```python
# First read the entire file into a string list
with open('/home/{}/testing/logs/contsrc_tint.log'.format(myusername)) as f:
    tmp_text=f.readlines()
f.close()

# Now write out each interval to a separate log file.
lstart=0
lend=nchan+1
times=[]
for t in range(nint):
    f=open('/home/{0}/testing/logs/t{1}.log'.format(myusername,str(t)),"w")
    for t in range(lstart,lend):
        f.write(tmp_text[t])
        if t==lstart:
            times.append(tmp_text[t].rstrip())
    f.close()
    lstart=lend
    lend+=nchan+1
```

Now we will convert the times into MJD to facilitate comparisons and labeling.

```python
from datetime import datetime
from astropy.time import Time
t_mjd=[]
for t in range(len(times)):
    t_mjd.append(Time(datetime.strptime(times[t].rstrip(), '%y%b%d:%H:%M:%S.%f')).mjd)
    
tzero=t_mjd[0]
for t in range(len(t_mjd)):
    t_mjd[t]=(t_mjd[t]-tzero)*24.
```

Okay, now we create a couple of 2D arrays so we can loop through the log file and save each spectrum to a separate row.

```python
# Create freq_t and amp_t arrays to hold each time interval in a separate row.  Initially they will be filled with 
# zeros.
freq_t=np.zeros([nchan,nint],float)
amp_t=np.zeros([nchan,nint],float)

# Now load the data from the log file into each row.
for t in range(nint):
    freq_t[:,t]=np.loadtxt('/home/{0}/testing/logs/t{1}.log'.format(myusername,str(t)),float,skiprows=1,usecols=[0])
    amp_t[:,t]=np.loadtxt('/home/{0}/testing/logs/t{1}.log'.format(myusername,str(t)),float,skiprows=1,usecols=[1])
```

Okay, now we can plot these spectra and see how they compare.

```python
offset=0.2
plt.plot(freq_avg,amp_avg-offset,'r-',label='Avg')
for t in range(nint):
    plt.plot(freq_t[:,t],amp_t[:,t]+t*offset,'-',label=round(t_mjd[t],2))
plt.xlim(np.min(freq_t[:,0])-0.01,np.max(freq_t[:,0])+0.1)
plt.xlabel('Frequency [GHz]')
plt.ylabel('Amplitude [Jy]')
plt.title('Continuum Source Flux at different times')
plt.legend()
plt.show()
```

Now we will divide each time interval by the average amplitudes and plot those

```python
# Normalize by average amplitude of continuum source to create ratio
amp_r=np.zeros([nchan,nint],float)

for t in range(nint):
    amp_r[:,t]=amp_t[:,t]/amp_avg[:]

# Calculate mean and stdev for each time interval
avg=[]
unc=[]
for t in range(nint):
    avg.append(np.mean(amp_r[:,t]))
    unc.append(np.std(amp_r[:,t]))    
    print("Time interval: ",t,round(t_mjd[t],2),"Mean ratio: ",avg[t],'+/-',unc[t])
    
# Plot each time interval
for t in range(nint):
    plt.plot(freq_t[:,t],amp_r[:,t],label=round(t_mjd[t],2))
plt.xlim(np.min(freq_t[:,0])-0.01,np.max(freq_t[:,0])+0.1)
plt.xlabel('Frequency [GHz]')
plt.ylabel('Amplitude [Jy]')
plt.legend()
plt.title('Ratio of continuum source flux to average value')
plt.show()

print("The RMS of the ratios (averaged over all channels) is: ",np.std(avg))
```

Finally, clean up scratch log files.

```python
import subprocess
os.system('rm -f /home/{}/testing/logs/t*.log'.format(myusername))
#os.listdir('/data/{0}/BP/'.format(myusername))
```

Okay, the bandpass stability specification is that there is less than a 1:10$^4$ variation across a 10 MHz frequency chunk over 12 hours.  We would like to see something similar for beam stability.  To test this directly, we will average the spectrum over 819 channels (approx. 10 MHz; 11 frequency chunks)

```python
df=12.20703125 # Channel width in kHz from a data cube
nchunk=int(nchan/round(10.e3/df))
freq_bin=np.zeros(nchunk,float)
ratio_bin=np.zeros([nchunk,nint],float)
rms_bin=np.zeros([nchunk,nint],float)

chanchunk=819 # The number of channels in a 10 MHz chunk (approx)
# Now we will derive the average frequency and average ratio for each 10 MHz chunk
c1=0
c2=chanchunk
for i in range(nchunk):
    freq_bin[i]=np.mean(freq_avg[c1:c2])
    for t in range(nint):
        ratio_bin[i,t]=np.mean(amp_r[c1:c2,t])
        rms_bin[i,t]=np.std(amp_r[c1:c2,t])
    c1+=chanchunk
    c2+=chanchunk
```

Okay, let's look at the ratio as a function of both frequency and time:

```python
for t in range(nint):
    plt.plot(freq_bin,ratio_bin[:,t])
plt.xlabel('Frequency [GHz]')
plt.ylabel('Flux ratio')
plt.show()

for t in range(nint):
    plt.plot(freq_bin,rms_bin[:,t])
plt.xlabel('Frequency [GHz]')
plt.ylabel('RMS of Flux ratio')
plt.show()

for c in range(nchunk):
    plt.plot(t_mjd,ratio_bin[c,:])
plt.xlabel('Time [hr]')
plt.ylabel('Flux ratio in 10 MHz chunk')
plt.show()
```

```python
# Calculate the RMS variation with time of each 10 MHz chunk
rms=[]
for c in range(nchunk):
    rms.append(np.std(ratio_bin[c,:]))
    print('RMS variation of ratio in frequency chunk ',c,' is ',rms[c])

print('Mean RMS is ',np.mean(rms))
print('The fraction of the goal (1:10000) is: ',np.mean(rms)/1e-4)
```

```python
# Just compare the absolute change from the first to the last time interval
for c in range(nchunk):
    print(freq_bin[c],abs(ratio_bin[c,nint-1]-ratio_bin[c,0]))
```

### Phase Stability Checks

Now that we have looked at amplitude stability, we should do a similar comparison for phases.  This comparison should be done by looking at offsets, however, and not fractional changes.  Otherwise the results should be the same.

```python
uvspec = lib.miriad('uvspec')

uvspec.vis = uvfile
uvspec.interval = tsolve  #Averaging the calibrator over 12 hours.
uvspec.stokes='ii'
uvspec.axis='frequency,phase'
uvspec.options='nobase,avall'
uvspec.log='/home/{}/testing/logs/contsrc_12h_phase.log'.format(myusername)
uvspec.offset=offsets

uvspec.go()
```

Now we will load the average bandpass into an array, plot it, and save it for reference. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

freq_avg=np.loadtxt('/home/{}/testing/logs/contsrc_12h_phase.log'.format(myusername),float,usecols=[0])
ph_avg=np.loadtxt('/home/{}/testing/logs/contsrc_12h_phase.log'.format(myusername),float,usecols=[1])
nchan=len(freq_avg)
plt.plot(freq_avg,ph_avg,'-')
plt.xlabel('Frequency [GHz]')
plt.ylabel('Phase [deg]')
plt.show()
```

Okay, now let's run uvspec with a shorter averaging interval and save the results to another log file, this time with a timestamp in it.  

```python
uvspec = lib.miriad('uvspec')

uvspec.vis = uvfile
uvspec.interval = tint  #Averaging the calibrator over tint.
uvspec.stokes='ii'
uvspec.axis='frequency,phase'
uvspec.options='nobase,avall,timelog'
uvspec.log='/home/{}/testing/logs/contsrc_tint_phase.log'.format(myusername)
uvspec.offset=offsets

uvspec.go()
```

```python
# First read the entire file into a string list
with open('/home/{}/testing/logs/contsrc_tint_phase.log'.format(myusername)) as f:
    tmp_text=f.readlines()
f.close()

# Now write out each interval to a separate log file.
lstart=0
lend=nchan+1
times=[]
for t in range(nint):
    f=open("/home/{0}/testing/logs/t{1}.log".format(myusername,str(t)),"w")
    for t in range(lstart,lend):
        f.write(tmp_text[t])
    f.close()
    lstart=lend
    lend+=nchan+1
```

```python
# Create freq_t and amp_t arrays to hold each time interval in a separate row.  Initially they will be filled with zeros.
ph_t=np.zeros([nchan,nint],float)

# Now load the data from the log file into each row.
for t in range(nint):
    ph_t[:,t]=np.loadtxt('/home/{0}/testing/logs/t{1}.log'.format(myusername,str(t)),float,skiprows=1,usecols=[1])
```

Okay, now we can plot these spectra and see how they compare.

```python
offset=0.2
plt.plot(freq_avg,ph_avg-offset,'r-',label='Avg')
for t in range(nint):
    plt.plot(freq_t[:,t],ph_t[:,t]+t*offset,'-',label=round(t_mjd[t],2))
plt.xlim(np.min(freq_t[:,0])-0.01,np.max(freq_t[:,0])+0.1)
plt.xlabel('Frequency [GHz]')
plt.ylabel('Phase [degrees]')
plt.title('Continuum Source Phase at different times')
plt.legend()
plt.show()
```

Now we will subtract the mean phases from each interval and plot that difference.

```python
# Normalize by average amplitude of Flux calibrator to create ratio
ph_r=np.zeros([nchan,nint],float)

for t in range(nint):
    ph_r[:,t]=ph_t[:,t]-ph_avg[:]

# Calculate mean and stdev for each time interval
avg=[]
unc=[]
for t in range(nint):
    avg.append(np.mean(ph_r[:,t]))
    unc.append(np.std(ph_r[:,t]))    
    print("Time interval: ",t,round(t_mjd[t],2),"Mean phase difference: ",avg[t],'+/-',unc[t])
    
# Plot each time interval
for t in range(nint):
    plt.plot(freq_t[:,t],ph_r[:,t],label=round(t_mjd[t],2))
plt.xlim(np.min(freq_t[:,0])-0.01,np.max(freq_t[:,0])+0.1)
plt.xlabel('Frequency [GHz]')
plt.ylabel('Phase difference [deg]')
plt.legend()
plt.title('Phase differences')
plt.show()

print("The RMS of the phase differences (averaged over all channels) is: ",np.std(avg))
```

```python
ph_ratio_bin=np.zeros([nchunk,nint],float)
ph_rms_bin=np.zeros([nchunk,nint],float)


# Now we will derive the average frequency and average difference for each 10 MHz chunk
c1=0
c2=chanchunk
for i in range(nchunk):
    for t in range(nint):
        ph_ratio_bin[i,t]=np.mean(ph_r[c1:c2,t])
        ph_rms_bin[i,t]=np.std(ph_r[c1:c2,t])
    c1+=chanchunk
    c2+=chanchunk
```

Okay, let's look at the phase difference as a function of both frequency and time:

```python
for t in range(nint):
    plt.plot(freq_bin,ph_ratio_bin[:,t])
plt.xlabel('Frequency [GHz]')
plt.ylabel('Phase difference')
plt.show()

for t in range(nint):
    plt.plot(freq_bin,ph_rms_bin[:,t])
plt.xlabel('Frequency [GHz]')
plt.ylabel('RMS Phase difference')
plt.show()

for c in range(nchunk):
    plt.plot(t_mjd,ph_ratio_bin[c,:])
plt.xlabel('Time [hr]')
plt.ylabel('Phase difference in 10 MHz chunk')
plt.show()
```

```python
# Calculate the RMS variation with time of each 10 MHz chunk
rms=[]
for c in range(nchunk):
    rms.append(np.std(ph_ratio_bin[c,:]))
    print('RMS variation of phase difference in frequency chunk ',c,' is ',rms[c])

print('Mean RMS is ',np.mean(rms))
```

```python
# Just compare the absolute change from the first to the last time interval
for c in range(nchunk):
    print(freq_bin[c],abs(ph_ratio_bin[c,nint-1]-ph_ratio_bin[c,0]))
```
